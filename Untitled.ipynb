```python
import torch
from torch import nn
import pytorch_lightning as pl
from modules import *
from hssm import *
import numpy as np
from matplotlib import pyplot as plt
from pytorch_lightning import loggers as pl_loggers
import pandas as pd
```

```python
parameters = {}
parameters['sampling_frequency'] = 240
parameters['training_set_size'] = 5000
parameters['spec_analysis_step'] = 10
parameters['num_processors'] = 5
parameters['num_components'] = 2
parameters['tsne_perplexity'] = 50
parameters['tsne_learning_rate'] = 200
parameters['tsne_max_iters'] = 300
parameters['tsne_exaggeration'] = 25

def gen_sine(dur, fs, f0):
    t = np.arange(dur)
    sinusoid = np.sin(2 * np.pi * t * (f0 / fs))

    return sinusoid


def create_simulated_data():
    np.random.seed(66)
    num_sections = np.random.randint(5000, 10000)
    num_features = 7
    num_behaviors = np.random.randint(20, 40)
    behavior_dict = {}
    for beh in range(num_behaviors):
        behavior_duration = np.random.randint(30, 50)
        features_dict = {}
        for feat in range(num_features):
            fs = parameters['sampling_frequency']
            f0 = np.random.randint(120, 360)
            sine1 = gen_sine(behavior_duration, f0, fs)
            f0 = np.random.randint(20, 120)
            sine2 = gen_sine(behavior_duration, f0, fs)
            features_dict[str(feat)] = sine1+sine2
        behavior_dict[str(beh)] = pd.DataFrame(features_dict)
    section_sequence = []
    behavior_sequence = []
    for sec in range(num_sections):
        behavior = np.random.randint(0, num_behaviors)
        section_sequence.append(behavior_dict[str(behavior)])
        behavior_sequence += [behavior] * len(behavior_dict[str(behavior)])

    return pd.concat(section_sequence).values, behavior_sequence

X, y = create_simulated_data()
X.shape
```

```python
seq_len = 300
n_samples = X.shape[0] // seq_len
X = X[:seq_len * n_samples].astype(np.float32)
dataset = X.reshape(n_samples, seq_len, -1)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, )
```

```python
class VTA_model(pl.LightningModule):
    def __init__(self):
        super(VTA_model, self).__init__()
        self.model = HierarchicalStateSpaceModel(belief_size=16,
                                                 input_size=7,
                                                 state_size=16,
                                                 num_layers=3,
                                                 max_seg_len=80,
                                                 max_seg_num=20)
        
    def forward(self, x):
        return self.model(x, seq_len-10, init_size=5)
    
    def training_step(self, batch, batch_idx):
        bx = batch
        model_output = self(bx)
        loss = loss_func(model_output, bx, init_size=5)
        train_loss = loss['train_loss']
        logs = {'loss': train_loss, 'obs_cost': loss['obs_cost'].mean()}
        return {'loss': train_loss, 'log': logs}
        
    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=0.0003)
    
    def train_dataloader(self):
        return torch.utils.data.DataLoader(dataset, batch_size=16, num_workers=4)
    
```

```python
%pdb on
model = VTA_model()
logger = pl_loggers.WandbLogger(project='VTA')
trainer = pl.Trainer(gpus=1, max_epochs=2, progress_bar_refresh_rate=10, logger=logger)
trainer.fit(model)
```
